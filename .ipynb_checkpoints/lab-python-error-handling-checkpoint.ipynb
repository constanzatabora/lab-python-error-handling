````python
#Exercise 2:
# Function to calculate the total price with error handling
def calculate_total_price(customer_orders):
    total_price = 0
    for product in customer_orders:
        valid_price = False
        while not valid_price:
            try:
                # Ask the user to enter the price of the product
                price = float(input(f"Enter the price of {product}: "))
                
                # Ensure the price is non-negative
                if price < 0:
                    raise ValueError("Price cannot be negative. Please enter a valid price.")
                
                # If the price is valid, add it to the total and break out of the loop
                total_price += price
                valid_price = True
                
            except ValueError as error:
                # Print the error message and prompt again
                print(f"Error: {error}")
    
    return total_price
````

````python
# Example customer_orders set
customer_orders = {'hat', 'keychain'}

# Calculate the total price with error handling
total_price = calculate_total_price(customer_orders)

# Print the total price
print(f"Total Price: {total_price:.2f}")
````

````python
#Exercise 3:
# Function to get customer orders with error handling
def get_customer_orders(inventory):
    customer_orders = set()

    # Step 1: Ask the user for the number of orders
    valid_number_of_orders = False
    while not valid_number_of_orders:
        try:
            # Prompt the user for the number of products to order
            number_of_orders = int(input("Enter the number of customer orders: "))
            if number_of_orders <= 0:
                raise ValueError("Number of orders must be a positive integer.")
            valid_number_of_orders = True
        except ValueError as error:
            # Handle invalid input
            print(f"Error: {error}")
    
    # Step 2: Ask the user to enter the product names
    for _ in range(number_of_orders):
        valid_product = False
        while not valid_product:
            try:
                # Ask the user to input the product name
                product = input("Enter the name of a product that a customer wants to order: ").lower()

                # Check if the product exists in inventory and has stock available
                if product not in inventory:
                    raise ValueError(f"{product} is not in the inventory. Please enter a valid product.")
                elif inventory[product] <= 0:
                    raise ValueError(f"{product} is out of stock. Please enter a different product.")
                
                # If valid, add the product to the customer orders
                customer_orders.add(product)
                valid_product = True
            except ValueError as error:
                # Handle invalid product input
                print(f"Error: {error}")

    return customer_orders
````

````python
# Example inventory dictionary
inventory = {
    't-shirt': 5,
    'mug': 4,
    'hat': 2,
    'book': 1,
    'keychain': 0  # No stock
}

# Get customer orders with error handling
customer_orders = get_customer_orders(inventory)

# Print customer orders
print("\nCustomer Orders:", customer_orders)
````

````python
#Exercise 4:
# Step 1: Function to initialize the inventory with error handling
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory

# Step 2: Function to get customer orders with error handling
def get_customer_orders(inventory):
    customer_orders = set()

    # Step 2a: Ask the user for the number of orders
    valid_number_of_orders = False
    while not valid_number_of_orders:
        try:
            number_of_orders = int(input("Enter the number of customer orders: "))
            if number_of_orders <= 0:
                raise ValueError("Number of orders must be a positive integer.")
            valid_number_of_orders = True
        except ValueError as error:
            print(f"Error: {error}")
    
    # Step 2b: Ask the user to enter the product names
    for _ in range(number_of_orders):
        valid_product = False
        while not valid_product:
            try:
                product = input("Enter the name of a product that a customer wants to order: ").lower()
                if product not in inventory:
                    raise ValueError(f"{product} is not in the inventory. Please enter a valid product.")
                elif inventory[product] <= 0:
                    raise ValueError(f"{product} is out of stock. Please enter a different product.")
                
                # If valid, add the product to the customer orders
                customer_orders.add(product)
                valid_product = True
            except ValueError as error:
                print(f"Error: {error}")

    return customer_orders

# Step 3: Test the functions by initializing the inventory and getting customer orders

# Example products list
products = ["t-shirt", "mug", "hat", "book", "keychain"]

# Initialize the inventory
inventory = initialize_inventory(products)

# Get customer orders
customer_orders = get_customer_orders(inventory)

# Display the customer orders
print("\nCustomer Orders:", customer_orders)
````
